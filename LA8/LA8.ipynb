```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('titanic.csv')
data
```

```python
data.shape
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
data['Age'] = data['Age'].fillna(np.mean(data['Age']))
data['Cabin'] = data['Cabin'].fillna(data['Cabin'].mode()[0])
data['Embarked'] = data['Embarked'].fillna(data['Embarked'].mode()[0])
data['Fare'] = data['Fare'].fillna(data['Fare'].mode()[0])
```

```python
data.isnull().sum()
```

```python
sns.countplot(x=data['Survived'])
```

```python
sns.countplot(x=data['Pclass'])
```

```python
sns.countplot(x=data['Embarked'])
```

```python
sns.countplot(x=data['Sex'])
```

```python
sns.boxplot(x=data['Age'])
```

```python
sns.boxplot(x=data['Fare'])
```

```python
sns.boxplot(x=data['SibSp'])
```

```python
sns.catplot(x= 'Pclass', y = 'Age', data=data, kind = 'box')
```

```python
sns.catplot(x= data['Pclass'], y = data['Age'] ,data=data,kind='bar')
```

```python
sns.catplot(x= 'Pclass', y = 'Fare', data=data, kind = 'strip')
```

```python
sns.catplot(x= 'Sex', y = 'Fare', data=data, kind = 'strip')
```

```python
sns.catplot(x= 'Sex', y = 'Age', data=data, kind = 'strip')
```

```python
sns.pairplot(data)
```

```python
sns.scatterplot(x = 'Fare', y = 'Pclass', hue = 'Survived', data = data)
```
